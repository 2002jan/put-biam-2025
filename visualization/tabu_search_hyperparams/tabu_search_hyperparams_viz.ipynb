```python
import numpy as np
import matplotlib.pyplot as plt

big_data = []
algos = []

with open("tabu_results_hyperparams.txt", 'r') as f:
    lines = f.readlines()

    i = 1

    while i < len(lines):

        params = lines[i].split(':')[1]
        i += 1
        params = params + lines[i].split(':')[1]
        i += 1
        algo = params.replace('\n', '').strip().replace(' ', '_').lower()
        algos.append(algo)

        # print(algo)

        data = []

        for _ in range(3):
            a = float(lines[i].replace('\n', '').split(':')[1].strip().split('(')[1].replace(')', ''))
            data.append(a)
            i += 1

        big_data.append(data)

        i += 1

values = np.array(big_data)
values
```

```python
values[:, 0]
```

```python

ordering = values[:, 2].argsort()[:-1]

sorted_algos = np.array(algos)[ordering]
sorted_values = values[ordering]
```

```python
fig = plt.figure(figsize=(12, 10))

ax = fig.add_subplot(1, 1, 1)

ax.barh(sorted_algos, sorted_values[:, 2], xerr=np.abs(sorted_values[:, 2] - np.array([sorted_values[:, 1], sorted_values[:, 0]])), capsize=5)
ax.set_xlim(1.15, 1.45)
```
